### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
random.seed(seedno)
#n=4x μέγεθος δικτύου: αντικαταστήστε το x με το τελευταίο ψηφίο του αριθμού μητρώου σας.
n = 47
d=4
# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
# Επιλέξτε προσεκτικά την τιμή της παραμέτρου p. 
#uncomment REG= 
p = 0
REG = nx.watts_strogatz_graph(n,d,p,seedno)

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG)
```

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)
#n=4x 
n = 47
M = 150

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
#uncomment RGER=
RGER = nx.gnm_random_graph(n,M,seed=seedno,directed = False)

# οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER)
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)
#n=4x
n = 47
R=0.35

# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
#uncomment RGG=
RGG = nx.random_geometric_graph(n,R,seed=seedno)

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html
```

```python
# Use seed when creating the graph for reproducibility
# G = nx.random_geometric_graph(200, 0.125, seed=896803)
# position is stored as node attribute data for random_geometric_graph
pos = nx.get_node_attributes(RGG, "pos")

# find node near center (0.5,0.5)
dmin = 1
ncenter = 0
for n in pos:
    x, y = pos[n]
    d = (x - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter = n
        dmin = d

# color by path length from node near center
p = dict(nx.single_source_shortest_path_length(RGG, ncenter))

plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(RGG, pos, alpha=0.4)
nx.draw_networkx_nodes(
    RGG,
    pos,
    nodelist=list(p.keys()),
    node_size=80,
    node_color=list(p.values()),
    cmap=plt.cm.Reds_r,
)

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis("off")
plt.show()
```

```python
import statistics

def cc(graph):
    # get clustering coefficient
    GraphCC = dict(nx.clustering(graph))
    
    # plot
    fig, ax = plt.subplots()
    ax.hist(GraphCC.values(), np.linspace(0,1, 10), color='magenta')
    ax.set_xticks([i/10 for i in range(11)])

    plt.title("Clustering Coefficient Distribution")
    plt.xlabel("Clustering Coefficient")
    plt.ylabel("#nodes")
    plt.show()
    
    # mean and var
    mmean = statistics.mean(list(GraphCC.values())) 
    vvar = statistics.variance(list(GraphCC.values()))
    print("Mean cc: " + str(mmean) + " | Var of cc: " + str(vvar))
    
    vals = list(GraphCC.values())
    mmax, mmin = vals.index(max(vals)), vals.index(min(vals)) 
    
    print("Node with maximum cc: " + str(mmax) + \
          " | node with minimum cc: " + str(mmin))
    print()

    
graphList = [(REG, 'REG'), (RGER, 'RGER'), (RGG, 'RGG'), (SF, 'SF'), (SW, 'SW')]

for (graph, name) in graphList:
    print(name+":")
    cc(graph)

```

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.
#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).
#### Για τα παραπάνω πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως παρουσιάζονται στη θεωρία.

```python
# returns mean value and variance for the elements in a Llist
import statistics

def calcMeanAndVar(Llist):
#     prob = dict()
#     for item in Llist:
#         if item in prob.keys():
#             prob[item]+=1
#         else:
#             prob[item] = 1
            
#     n = len(Llist)        
#     for key in prob.keys():
#         prob[key] = prob[key]/n
# #     print(prob.items())
        
#     mean = sum([item[0]*item[1] for item in prob.items()])
    
#     var = sum(item[1]*((item[0]-mean)**2) for item in prob.items())
    
    return statistics.mean(Llist), statistics.variance(Llist)
    
```

```python
def compute_path_lengths(G):
    pathDict = nx.all_pairs_shortest_path_length(G)
    pathLengths = []
        
    for (source, destinationsDict) in pathDict:
        for (dest, distance) in destinationsDict.items():
            pathLengths.append(distance)
    
    mmean, vvar = calcMeanAndVar(pathLengths)
    # if G is connected calculate the diameter and the radius 
    if nx.is_connected(G):
        diam, rad = nx.diameter(G), nx.radius(G)
    # else both the diameter and the radius are infinite
    else:
        diam, rad = float('inf'), float('inf')
    
    return mmean, vvar, diam, rad
```

```python
graphList = [(REG, 'REG'), (RGER, 'RGER'), (RGG, 'RGG'), (SF, 'SF'), (SW, 'SW')]

for (graph, name) in graphList:
    print(name + ":")
    mmean, vvar, diam, rad = compute_path_lengths(graph)  
    print("Mean Distance: " + str(mmean) + " | Variance of Distance: " + str(vvar)) 
    print("Diameter: " + str(diam) + " | Radius: " + str(rad))
    print()
```

#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
def find_nodes(G):
    best_source,  best_destinations, best = 0, [], 0
    pathDict = dict(nx.all_pairs_shortest_path_length(G))
    
    if nx.is_connected(G):
        diameter = nx.diameter(G)
        
        for (source, destinationsDict) in pathDict.items():
            destinations, count = [], 0
            for (dest, distance) in destinationsDict.items():
                if distance == diameter:
                    count += 1
                    destinations.append(dest)
#                     print(source, dest, distance)
            if count > best:
                best_source, best_destinations, best = source, destinations, count
        print("diameter: " + str(diameter))
        
    else:
        #if the graph is not connected find the most isolated node
        for (source, destinationsDict) in pathDict.items():
            if len(destinationsDict) > best:
                best_source, best = source, len(destinationsDict)
                best_destinations = list(set(pathDict.keys())-set(destinationsDict.keys()))
                
    
    return best_source, best_destinations
                
```

```python
# να προσδιοριστεί το label του κόμβου και το σύνολο των κόμβων 
# με τους οποίους συνδέεται με μήκος ελάχιστου μονοπατιού ΄ίσο με τη διάμετρο. 
graphList = [(REG, 'REG'), (RGER, 'RGER'), (RGG, 'RGG'), (SF, 'SF'), (SW, 'SW')]

for (graph, name) in graphList:
    print(name + ":")
    nodeLabel, nodeList = find_nodes(graph)
    print("The requested node: " + str(nodeLabel))
    print("The list of nodes that are connected to node "+ str(nodeLabel) + " with paths of length equal to the diameter: ")
    print(nodeList)
    print()
```

#### Περιφέρεια του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με τη διάμετρό του. 
#### Κέντρο του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με την ακτίνα του.  
#### Β.2.4 Για κάθε τοπολογία, να βρεθεί η περιφέρεια και το κέντρο της και να σχολιαστούν τα αποτελέσματα.

```python
def find_periphery(G):
    pathDict = dict(nx.all_pairs_shortest_path_length(G))
    per = []
    
    if nx.is_connected(G):
        diameter = nx.diameter(G)
        
        for (source, destinationsDict) in pathDict.items():
            distances = destinationsDict.values()
            if max(distances) == diameter:
                per.append(source)
        
    else:
        #if the graph is not connected all nodes are in the periphery (trivial)
        per = list(G.nodes)             
    
    return per
```

```python
def find_center(G):
    pathDict = dict(nx.all_pairs_shortest_path_length(G))
    cen = []
    
    if nx.is_connected(G):
        radius = nx.radius(G)
        
        for (source, destinationsDict) in pathDict.items():
            distances = destinationsDict.values()
            if max(distances) == radius:
                cen.append(source)
        
    else:
        #if the graph is not connected all nodes are in the center (trivial)
        cen = list(G.nodes)             
    
    return cen
```

```python
graphList = [(REG, 'REG'), (RGER, 'RGER'), (RGG, 'RGG'), (SF, 'SF'), (SW, 'SW')]

for (graph, name) in graphList:
    print(name + ":")
    per, cen = find_periphery(graph), find_center(graph)
    if not nx.is_connected(graph):
        print("(the graph is not connected, therefore trivially all nodes belong to the periphery and the center)")
    print("Periphery: ", per)
    print("Center: ", cen)
    print()
```

## Γ) Mετρικές κεντρικότητας κόμβων

### Γ1. Κεντρικότητα βαθμού (Degree centrality)
##### Γ1.1 Να αναπαρασταθεί η κατατανομή βαθμών κόμβων για κάθε τοπολογία.
##### Γ1.2 Να υπολογιστεί ο μέσος βαθμός κόμβου καθώς και η διασπορά των βαθμών κόμβου για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_degree(G):
    degs = G.degree
    degrees = [b for (a,b) in degs]
    plt.hist(degrees, color='magenta')
    plt.title("Distribution of Degrees")
    plt.show()
    
    return statistics.mean(degrees), statistics.variance(degrees)
    
```

```python
graphList = [(REG, 'REG'), (RGER, 'RGER'), (RGG, 'RGG'), (SF, 'SF'), (SW, 'SW')]

for (graph, name) in graphList:
    print(name + ":")
    mmean, vvar = find_degree(graph)
    print("Mean Degree: " + str(mmean) + " | Variance of Degree: " + str(vvar)) 
    print()
```

### Γ2. Κεντρικότητα εγγύτητας (Closeness centrality)
##### Γ2.1 Να αναπαρασταθεί η κατατανομή της εγγύτητας κόμβων για κάθε τοπολογία.
##### Γ2.2 Να υπολογιστεί η μέση εγγύτητα καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_closeness(G):
    close_cen = nx.closeness_centrality(G)
    close_cen_values = [round(item, 10) for item in close_cen.values()]
    plt.hist(close_cen_values, color='magenta')
    plt.title("Distribution of Closeness Centrality")
    plt.show()
    return statistics.mean(close_cen_values), \
            statistics.variance(close_cen_values)
```

```python
graphList = [(REG, 'REG'), (RGER, 'RGER'), (RGG, 'RGG'), (SF, 'SF'), (SW, 'SW')]

for (graph, name) in graphList:
    print(name + ":")
    mmean, vvar = find_closeness(graph)
    print("Mean Closeness: " + str(mmean) + " | Variance of Closeness: " + str(vvar)) 
    print()
```

##### Γ3.2 Να υπολογιστεί η μέση ενδιαμεσική κεντρικότητα καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_betweenness(G):
    bet_cen = nx.betweenness_centrality(G)
    bet_cen_values = [round(item,10) for item in bet_cen.values()]
    fig, ax = plt.subplots()
#     mmax = max(bet_cen_values)
#     mmin = min(bet_cen_values)
    plt.hist(bet_cen_values, color='magenta')
#     ax.set_xticks(np.linspace(0,0.2,10))
    plt.title("Distribution of Betweenness Centrality")
    plt.show()
    return statistics.mean(bet_cen_values), \
            statistics.variance(bet_cen_values)
```

### Γ4. Kεντρικότητα Katz
##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.
##### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_betweenness(G):
    bet_cen = nx.betweenness_centrality(G)
    bet_cen_values = [round(item,10) for item in bet_cen.values()]
    fig, ax = plt.subplots()
#     mmax = max(bet_cen_values)
#     mmin = min(bet_cen_values)
    plt.hist(bet_cen_values, color='magenta')
#     ax.set_xticks(np.linspace(0,0.2,10))
    plt.title("Distribution of Betweenness Centrality")
    plt.show()
    return statistics.mean(bet_cen_values), \
            statistics.variance(bet_cen_values)


def find_katz(G):
    # adj spectrum 
    adj = nx.adjacency_spectrum(G)
    # print(adj)
    max_eigen = np.real(max(adj))
    al = (1/max_eigen)-0.01
    
    katz_cen = nx.katz_centrality(G,alpha=al)
    katz_cen_values = [round(item,10) for item in katz_cen.values()]
    
    plt.hist(katz_cen_values, color='magenta')
#     ax.set_xticks(np.linspace(0,0.2,10))
    plt.title("Distribution of Katz Centrality")
    plt.show()
    return statistics.mean(katz_cen_values), \
            statistics.variance(katz_cen_values)
```

```python
graphList = [(REG, 'REG'), (RGER, 'RGER'), (RGG, 'RGG'), (SF, 'SF'), (SW, 'SW')]

for (graph, name) in graphList:
    print(name + ":")
    mmean, vvar = find_katz(graph)
    print("Mean Katz Centrality: " + str(mmean) + " | Variance of Katz Centrality: " + str(vvar)) 
    print()
```

#### Να προσδιοριστούν, για κάθε τοπολογία, οι πιο κεντρικοί κόμβοι ως προς το βαθμό, την εγγύτητα, την ενδιαμεσική κεντρικότητα και την κεντρικότητα Katz. Να σχολιαστούν τα αποτελέσματα.

```python
def find_central(G):
    # degrees
    degs = G.degree
    degrees = [b for (a,b) in degs]
    max_deg, max_deg_nodes = max(degrees), []
    for (node,deg) in degs:
        if deg == max_deg:
            max_deg_nodes.append(node)
            
    print('nodes with maximum degree: ')
    print(max_deg_nodes)
    
    # closeness
    close_cen = nx.closeness_centrality(G)
    close_cen_values = [round(item, 12) for item in close_cen.values()]
    max_closeness, max_closeness_nodes = max(close_cen_values), []
    for (node, cen) in close_cen.items():
        if round(cen,12) == max_closeness:
            max_closeness_nodes.append(node)
            
    print('nodes with maximum closeness centrality: ')
    print(max_closeness_nodes)
    
    # betweenness
    bet_cen = nx.betweenness_centrality(G)
    bet_cen_values = [round(item,12) for item in bet_cen.values()]
    max_bet, max_bet_nodes = max(bet_cen_values), []
    for (node, bet) in bet_cen.items():
        if round(bet,12) == max_bet:
            max_bet_nodes.append(node)
    
    print('nodes with maximum betweenness centrality: ')
    print(max_bet_nodes)
    
    # Katz
    adj = nx.adjacency_spectrum(G)
    max_eigen = np.real(max(adj))
    al = (1/max_eigen)-0.01
    katz_cen = nx.katz_centrality(G,alpha=al)
    katz_cen_values = [round(item,12) for item in katz_cen.values()]
    max_katz, max_katz_nodes = max(katz_cen_values), []
    for (node, katz) in katz_cen.items():
        if round(katz,12) == max_katz:
            max_katz_nodes.append(node)
    
    print('nodes with maximum katz centrality: ')
    print(max_katz_nodes)
    
    
    print()
```

```python
graphList = [(REG, 'REG'), (RGER, 'RGER'), (RGG, 'RGG'), (SF, 'SF'), (SW, 'SW')]

for (graph, name) in graphList:
    print(name + ":")
    find_central(graph)
```
